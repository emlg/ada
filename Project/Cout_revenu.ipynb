```python
import pandas as pd
import px_reader as px
from helpers import *
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cout_revenu = read_file('cout-revenu.px')
cout_revenu.head()
```

```python
cout_revenu = cout_revenu.reset_index().rename(columns = {'level_0': 'annee','level_1': 'domaine',
                                                 'level_2': 'poste', 'level_3' : 'fond'})
cout_revenu.head()
```

```python
cout_revenu.columns[4:]
```

```python
cout_revenu = parse_df(cout_revenu, 4)
```

```python
cout_revenu = cout_revenu.groupby(['annee', 'poste', 'fond']).sum()
cout_revenu.head()
```

```python
type_postes = list(cout_revenu.index.get_level_values('poste').unique())
type_fonds = list(cout_revenu.index.get_level_values('fond').unique())
years = list(cout_revenu.index.get_level_values('annee').unique())
```

```python
type_postes
```

```python
def get_df_for_uni(uni, summarized = True):
    uni_df = pd.DataFrame(columns = type_postes)
    uni_df.insert(loc=0, column='annee', value=years)
    uni_df = uni_df.set_index('annee')
    tmp_df = cout_revenu[[uni]]
    if summarized:
        tmp_df = tmp_df.groupby(['annee', 'poste']).sum()
        for y in years:
            uni_df.loc[y] = list((tmp_df.loc[y][uni].reset_index())[uni])
        uni_df['Enseignement'] = uni_df[['Enseignement : etudes de base',
 'Enseignement : formation approfondie']].sum(axis=1)
        return uni_df
    else:
        return tmp_df
```

```python
summ_EPFL = get_df_for_uni('EPFL')
summ_EPFL
```

```python
summ_EPFL.plot(kind = 'line', logy = False)
```
